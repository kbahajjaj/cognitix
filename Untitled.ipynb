```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create a sample DataFrame with categorical data
df = pd.DataFrame({
    "colour": ["red", "blue", "green", "red", "red", "yellow", "blue"],
    "direction": ["up", "up", "down", "left", "right", "down", "down"],
    "numeric_col": [1, 2, 3, 4, 5, 6, 7] # Include a numeric col to show how to ignore it
})

# 2. Identify categorical columns
# You can manually list them or select dtypes 'category' or 'object'
categorical_features = ["colour", "direction"] 
# Alternative: 
# categorical_features = df.select_dtypes(include=['object', 'category']).columns.tolist()

# 3. Create subplots for each categorical feature
fig, ax = plt.subplots(1, len(categorical_features), figsize=(12, 5)) # Adjust figsize as needed
print(ax)

# 4. Loop through columns, calculate value counts, and plot
for i, feature in enumerate(categorical_features):
    print(feature)
    print(ax)
    print(type(ax))
    # Calculate the frequency of each category
    counts = df[feature].value_counts()
    
    # Plot the bar chart on the corresponding subplot axis
    counts.plot(kind="bar", ax=ax[i], title=feature)
    
    # Set axis labels for clarity
    ax[i].set_xlabel(feature)
    ax[i].set_ylabel("Frequency")
    
    # Rotate x-axis labels if needed for better readability
    ax[i].set_xticklabels(counts.index, rotation=45, ha='right')

# Adjust layout to prevent title/label overlap
plt.tight_layout() 
plt.show()
```
