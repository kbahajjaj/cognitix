## 2.2 Checking which columns/rows to drop and which to fix:
There are 38 columns which is a lot of features

### 2.2.1 Info vs Nunique vs Unique Values vs Nulls Percentage
Vieing info beside N-Unique values and unique values for better comparison and further inspection before final decision

```python
def info_plus(df):
    info_nunique = [[k, df[k].notnull().sum(), df[k].isnull().sum(), df[k].dtypes, df[k].nunique(), (df[k].unique().tolist()[:4]),\
                    int(df[k].isnull().sum()/len(df[k])*100)] for k in df.columns]
    info_nunique_df = pd.DataFrame(info_nunique, columns=['Column', 'Non-Null', 'Nulls', 'DType', 'N Unique', 'First 5 Unique', '% Missing'])
    return info_nunique_df

df_info = info_plus(df)
df_info
```

```python
df_info[df_info['Nulls'] > 0]
```

# 3. Clearning Data
**Columns/Rows to Modify or Drop:**

    - City (vital) -> Keep
4. Columns with too many missing values: offers, churn category, churn reason -> Drop
5. Related columns (highly correlated): Monthly charge, total charges, total refunds, total extra data charges, total long distance charges, total revenue -> choose one of them
6. Rows or data entries not related to churn/not churn: Customer status "Joined" rows -> Drop rows if small % or turn into "Stayed" if more than 5%

## 3.3 Column - Customer Status: Replace "Joined" with "Stayed"

```python
print(f'"Joined"% in Cusomer Status = %{(df["Customer Status"] == "Joined").sum().item()/df.shape[0]*100}')
print('Can either be ropped or changed to "Stayed", we will choose to change it as a joined client is considered a stayed one')
```
